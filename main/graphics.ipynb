```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datapoints = pd.read_csv('data/bonds_complete.csv')
datapoints = datapoints.rename(columns={'Unnamed: 0': 'tser'})
datapoints['dates'] = pd.to_datetime(datapoints['dates'])
datapoints.sort_values(by='dates')
```

```python
datapoints['change'] = ((datapoints['closing_values'] - datapoints['opening_values'])/datapoints['opening_values'])
```

```python
plt.plot(datapoints['dates'],datapoints['change'], color='b')
```

```python
plt.plot(datapoints['dates'],datapoints['opening_values'], color='r')
```
